```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
x,y = make_regression(n_samples=4 ,n_features=1,n_informative=1,n_targets=1,noise=80,random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
# Let's apply OLS 
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(x,y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler

# Create a scaler object
scaler = StandardScaler()

# Scale the features in your training data (x_train)
x_train_scaled = scaler.fit_transform(x_train)

# Create the LinearRegression model without normalize
lr = LinearRegression()

# Train the model using the scaled features
lr.fit(x_train_scaled, y_train)
```

```python
# Let's apply Gradient Descent assuming slope is constant m=78.35
# and Let's assume the starting value for intercept b=0
y_pred = ((78.35* x) + 100).reshape(4)
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red',label='OLS')
plt.plot(x,y_pred,color='#00a65a',label='b=0')
plt.legend()
plt.show()
```

```python
m=78.35
b=100

loss_slope = -2 * np.sum(y-m*x.ravel()-b)
loss_slope
```

```python
# Let's take learning rate =0.1
lr=0.1

step_size = loss_slope*lr
step_size
```

```python
# Calculating the new intercept
b = b - step_size
b
```

```python
y_pred1 = ((78.35*x) + b).reshape(4)

plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red',label='OLS')
plt.plot(x,y_pred1,color='#00a65a',label='b = {}'.format(b))
plt.plot(x,y_pred,color='#A3E4D7',label='b = 0')
plt.legend()
plt.show()
```
