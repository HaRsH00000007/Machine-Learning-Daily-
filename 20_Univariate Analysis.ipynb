# Categorical Data

# a.Countplot

```python
# How many of them survived.
#sns.countplot(df['Survived'])
df['Survived'].value_counts().plot(kind='bar')
```

```python
# How many of them were travelling in which class
#sns.countplot(df['Pclass'])
df['Pclass'].value_counts().plot(kind='bar')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Gender Distribution
sns.countplot(x='Sex', data=df)
plt.title('Gender Distribution')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()

```

```python
sns.countplot(df['Sex'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plotting count of passengers by embarked port
sns.countplot(x='Embarked', data=df)
plt.title('Passenger Count by Embarked Port')
plt.xlabel('Embarked Port')
plt.ylabel('Count')
plt.show()
```


# b. PieChart

```python
# How many of them survived showing percentage.
df['Survived'].value_counts().plot(kind='pie',autopct='%.2f')
```

```python
# How many of them were travelling in which class showing percentage.
df['Pclass'].value_counts().plot(kind='pie',autopct='%.2f')
```

```python
# Gender Distribution showing percentage
df['Sex'].value_counts().plot(kind='pie',autopct='%.2f')
```
